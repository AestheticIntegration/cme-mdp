```python
cp = []
dic = {"tests":[]}
for f in select:
    base = f.split("/")[-1]
    code = base.split(".")[0]
    code = code.split("_")[1:]
    entry = {}
    entry["file"] = "_".join(code)
    code[0] = code[0][:5]
    entry["name"] = "_".join(code)
    dic["tests"].append(entry)
```

```python
"{0}_{1:04}".format(*dic["tests"][0]["name"].split("_"))
```

```python
def k(x):
    s,v = x["name"].split("_")
    return (s,int(v))
dic["tests"] = sorted(dic["tests"], key=k)
```
